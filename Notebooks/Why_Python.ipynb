###**2-2-4. Dictionaries**

````python
# A dictionary stores KEY → VALUE pairs.
# Keys are usually strings (but can be numbers, tuples, etc.).
# Values can be anything (numbers, strings, lists, even other dicts).

# Create a simple dictionary: material → band gap (eV)
band_gaps = {
    "Si": 1.12,
    "GaAs": 1.42,
    "Ge": 0.66
}

print("Dictionary:", band_gaps)

# Access a value by its key
print("Band gap of Si:", band_gaps["Si"])

# Add a new key–value pair
band_gaps["GaN"] = 3.4
print("After adding GaN:", band_gaps)

# Modify an existing value
band_gaps["Ge"] = 0.70
print("After updating Ge:", band_gaps)

# Check if a key exists
print("Is 'Si' in dictionary?", "Si" in band_gaps)
print("Is 'Fe2O3' in dictionary?", "Fe2O3" in band_gaps)
````

````python
# Loop over KEYS
print("Materials:")
for mat in band_gaps.keys():
    print(" -", mat)

# Loop over VALUES
print("\nBand gaps (eV):")
for gap in band_gaps.values():
    print(" -", gap)

# Loop over (KEY, VALUE) pairs
print("\nMaterial → band gap:")
for mat, gap in band_gaps.items():
    print(f" - {mat}: {gap} eV")
````

````python
# Nested dictionaries: each material has several properties
materials = {
    "Si":  {"band_gap": 1.12, "structure": "diamond", "Z": 14},
    "GaAs": {"band_gap": 1.42, "structure": "zincblende", "Z_avg": 32.5},
    "Al2O3": {"band_gap": 8.8, "structure": "corundum", "Z_avg": 13}
}

# Access one material's properties
print("Si properties:", materials["Si"])
print("Si band gap:", materials["Si"]["band_gap"])

# Loop through all materials and print a summary
for name, props in materials.items():
    print(f"{name}: band gap = {props['band_gap']} eV, structure = {props['structure']}")
````

###**2-2-6. Classes:**

In Python, everything has a type (int, float, str, list, …).
A class is how we define our own type.

A class describes:

* what data an object has (attributes)
* what it can do (methods)



````python
# Define a new type: Point in 2D
class Point:
    # __init__ is the CONSTRUCTOR: it runs when we create a new Point
    def __init__(self, x, y):
        self.x = x  # attribute x
        self.y = y  # attribute y

    # A METHOD: function that belongs to Point objects
    def move(self, dx, dy):
        self.x += dx
        self.y += dy

    def distance_to_origin(self):
        return (self.x**2 + self.y**2) ** 0.5


# Create two points (objects / instances)
p1 = Point(0, 0)
p2 = Point(3, 4)

print("p2.x:", p2.x, "p2.y:", p2.y)

# Call methods
p2.move(-1, 2)
print("After move → p2:", p2.x, p2.y)

print("Distance of p2 to origin:", p2.distance_to_origin())
````

````python
import matplotlib.pyplot as plt

# Create a point and move it
p = Point(1, 2)
print("Original:", p.x, p.y)

# Save original coordinates
x0, y0 = p.x, p.y

# Move the point
p.move(2, -1)
print("After move:", p.x, p.y)

# Plot original and new position
plt.figure(figsize=(4, 4))
plt.scatter([x0], [y0], color="blue", label="Original")
plt.scatter([p.x], [p.y], color="red", label="After move")

# Draw arrow showing movement
plt.arrow(x0, y0, p.x - x0, p.y - y0,
          length_includes_head=True,
          head_width=0.1, color="gray")

plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")
plt.show()
````

````python
import random

points = [Point(random.uniform(-3, 3), random.uniform(-3, 3)) for _ in range(5)]

plt.figure(figsize=(4, 4))

for p in points:
    x0, y0 = p.x, p.y
    dx, dy = random.uniform(-1, 1), random.uniform(-1, 1)
    p.move(dx, dy)
    plt.scatter([x0], [y0], color="blue")
    plt.scatter([p.x], [p.y], color="red")
    plt.arrow(x0, y0, p.x - x0, p.y - y0,
              length_includes_head=True,
              head_width=0.07, color="gray")

plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")
plt.show()
````

````python
class Material:
    def __init__(self, name, formula, band_gap_ev):
        self.name = name
        self.formula = formula
        self.band_gap_ev = band_gap_ev

    def is_metal(self):
        # Very rough rule: band gap ~ 0 → metal
        return self.band_gap_ev < 0.1

    def describe(self):
        kind = "metal" if self.is_metal() else "insulator/semiconductor"
        print(f"{self.name} ({self.formula}): band gap = {self.band_gap_ev} eV → {kind}")


si = Material("Silicon", "Si", 1.12)
cu = Material("Copper", "Cu", 0.0)

si.describe()
cu.describe()

print("Is Si a metal?", si.is_metal())
print("Is Cu a metal?", cu.is_metal())
````

###**2-3-3. SciPy: scientific computing toolbox on top of NumPy**


SciPy is a collection of numerical algorithms built on top of NumPy.  
Where NumPy gives you fast arrays and linear algebra, SciPy adds:

- **Optimization** (minimization, curve fitting): `scipy.optimize`
- **Integration and ODE solvers**: `scipy.integrate`
- **Interpolation / gridding** of data: `scipy.interpolate`
- **Signal/image processing**: `scipy.signal`, `scipy.ndimage`
- **Statistics**: `scipy.stats`

In materials science, SciPy is typically used to:
- Fit **material parameters** (e.g. diffusion coefficients, kinetic rates, constitutive model parameters) to experimental/simulation data.
- Solve **kinetic ODEs** (precipitation, transformation, oxidation).
- Interpolate / smooth **fields** (e.g. stress, composition) from simulations or experiments onto regular grids.


````python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Synthetic 1D diffusion profile: C(x,t) ~ error function
# C(x) = C0 + (C1 - C0) * erf( x / (2*sqrt(D*t)) )

def erfc_profile(x, D, C0, C1, t):
    from math import sqrt
    from scipy.special import erf
    return C0 + (C1 - C0) * erf(x / (2 * np.sqrt(D * t)))

# "Experimental" data (here we just generate some)
np.random.seed(0)
x = np.linspace(0, 1e-3, 50)    # 0..1 mm
D_true, C0_true, C1_true, t = 1e-14, 0.1, 1.0, 3600.0  # example
C = erfc_profile(x, D_true, C0_true, C1_true, t) + 0.02*np.random.randn(len(x))

# Fit D, C0, C1 from data
def model(x, D, C0, C1):
    return erfc_profile(x, D, C0, C1, t)

popt, pcov = curve_fit(model, x, C, p0=[1e-15, 0.0, 1.0])
D_fit, C0_fit, C1_fit = popt
print("Fitted D:", D_fit)

# Plot
x_fit = np.linspace(0, 1e-3, 200)
plt.scatter(x, C, s=10, label="data")
plt.plot(x_fit, model(x_fit, *popt), 'r-', label="fit")
plt.xlabel("x (m)")
plt.ylabel("C")
plt.legend()
plt.show()
````

````python

from scipy.integrate import solve_ivp

# Simple JMAK-like model: dX/dt = k * (1 - X)^n
# X(t): transformed fraction (e.g. precipitated volume fraction)

k = 1e-3   # 1/s
n = 2.0

def dXdt(t, X):
    return k * (1.0 - X)**n

t_span = (0.0, 2e4)      # 0 .. 20000 s
t_eval = np.linspace(*t_span, 200)

sol = solve_ivp(dXdt, t_span, y0=[0.0], t_eval=t_eval)
X = sol.y[0]

plt.plot(t_eval/3600, X)
plt.xlabel("time (h)")
plt.ylabel("transformed fraction X")
plt.title("Simple precipitation / transformation kinetics")
plt.grid(True)
plt.show()
````

````python
from scipy.interpolate import griddata


# Suppose we measured hardness at scattered (x,y) positions
np.random.seed(0)
N = 200
x_sc = np.random.uniform(0, 1, N)
y_sc = np.random.uniform(0, 1, N)
# synthetic "true" field: hardness = 200 + 50*x + 30*y
H_sc = 200 + 50*x_sc + 30*y_sc + 5*np.random.randn(N)

points = np.column_stack([x_sc, y_sc])

# Regular grid
nx = ny = 64
grid_x, grid_y = np.meshgrid(
    np.linspace(0, 1, nx),
    np.linspace(0, 1, ny)
)

# Interpolate scattered data onto grid
H_grid = griddata(points, H_sc, (grid_x, grid_y), method='cubic')

plt.figure(figsize=(6, 4))
im = plt.imshow(H_grid, origin='lower', extent=(0,1,0,1), cmap='viridis')
plt.scatter(x_sc, y_sc, s=5, c='k', alpha=0.5, label="data points")
plt.colorbar(im, label="Hardness")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolated hardness map (SciPy interpolate)")
plt.legend()
plt.show()
````
